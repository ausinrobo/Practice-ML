```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

train_data = pd.read_excel('Task1and2/train.xlsx')

features = train_data[['feature1', 'feature2', 'feature3']]  
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)  
kmeans.fit(scaled_features)
clusters = kmeans.predict(scaled_features)
train_data['Cluster'] = clusters

def classify_new_data_point(new_data_point):
    scaled_data_point = scaler.transform([new_data_point])
    cluster = kmeans.predict(scaled_data_point)
    return cluster[0]


import numpy as np

def explain_cluster_membership(data_point, cluster):
    scaled_data_point = scaler.transform([data_point])
    centroid = kmeans.cluster_centers_[cluster]
    distance = np.linalg.norm(scaled_data_point - centroid)
    return f"Data point {data_point} is in cluster {cluster} because it is closest to the centroid {centroid} with a distance of {distance}"

explanation = explain_cluster_membership(new_data_point, cluster)
print(explanation)

```
